## Process the dataset

```python
airbnb_df.dropna()
```

```python
plt.hist(airbnb_df['reviews_per_month'], bins=100);
```

```python
airbnb_df.drop(columns=['reviews_per_month'], inplace=True)
```

```python
airbnb_df.drop(columns=['neighbourhood'], inplace=True)
```

```python
airbnb_df.isna().sum()
```

```python
airbnb_df.head()
```

```python
airbnb_df = pd.get_dummies(airbnb_df, dtype = 'int', drop_first=True)
```

```python
airbnb_df
```

```python
from sklearn.model_selection import train_test_split

```

```python
train_set, test_set = train_test_split(airbnb_df)
```

```python
test_set
```

#### Let's save the Datasets

```python
train_set.to_csv('../data/processed/trainset.csv', index=False)
test_set.to_csv('../data/processed/testset.csv', index=False)
```
